```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Retina display
%config InlineBackend.figure_format = 'retina'
```

```python
# Simple data for SVD in two dimensions
x = np.array([[0, 1], [1, 0], [0, -1], [-1, 0]])

# Make an aray of colours for the points
colours = 'green'

# use a dark background
plt.style.use('dark_background')

# Plot a scatter plot of the points using colours from the array
plt.plot(x[:, 0], x[:, 1], 'o', c=colours, markersize=10)

# Set equal aspect ratio
plt.gca().set_aspect('equal')

# Draw a circle joining the points
t = np.linspace(0, 2*np.pi, 100)
plt.plot(np.cos(t), np.sin(t), 'w--')

# Fill the circle
plt.fill(np.cos(t), np.sin(t), 'w', alpha=0.1)

# Make the plot look nicer
plt.tight_layout()
```

```python
# Find the SVD

U, s, V = np.linalg.svd(x, full_matrices=False)
```
